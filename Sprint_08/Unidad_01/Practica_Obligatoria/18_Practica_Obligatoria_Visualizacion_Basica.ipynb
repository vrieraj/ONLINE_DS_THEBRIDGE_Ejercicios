1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)

1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)

```python
df_top_pop = df_casas.nlargest(5,'population_total').reset_index()
df_top_pop.population_total = df_top_pop.population_total / 1000
colores = ['blue','grey','red','olive','brown']

plt.bar(df_top_pop.city, height = df_top_pop.population_total, color = colores, alpha=0.5)
for i in range(len(df_top_pop)):
    plt.text(i, (df_top_pop.population_total[i]) +100, df_top_pop.population_total[i], ha = 'center')
plt.ylim(0,5000)
plt.ylabel('Habitantes (miles)')
plt.xlabel('Ciudades')
plt.title('Ciudades más pobladas de California');
```

2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)

```python
plt.hist(df_casas.area_total_km2, bins = 50, color = 'steelblue', label = 'Hist Area')
fuente = {'font':'Bradley Hand itc','size':18}
plt.xlabel('Superficie ($km^2$)',fontdict=fuente)
plt.ylabel('Número de ciudades',fontdict=fuente)
plt.legend();
```

3. El scatter de área de agua de un ciudad con su poblacion.

```python
outliers = df_casas.nlargest(2,'area_water_km2')
coord_outliers = []
for ciudad in outliers.city:
    x = df_casas.loc[df_casas.city == ciudad,'area_water_km2']
    y = df_casas.loc[df_casas.city == ciudad,'population_total']/1000
    coord_outliers.append((float(x),float(y)))
coord_outliers
```

```python
plt.scatter(df_casas.area_water_km2, df_casas.population_total / 1000, color = 'g', alpha=0.5)
plt.xlabel('Área de agua en la ciudad ($km^2$)')
plt.ylabel('Población (miles habitantes)')
plt.title('Población por kilómetro cuadrado de agua')
coordtext = (60,3500)
plt.annotate('Outliers',xy = coord_outliers[0], xytext=coordtext, arrowprops=dict(arrowstyle = '->'))
plt.annotate('Outliers',xy = coord_outliers[1], xytext=coordtext, arrowprops=dict(arrowstyle = '->'));
```

4. El boxplot de la altura de las ciudades.

```python
plt.boxplot(df_casas.elevation_m, orientation = 'horizontal', patch_artist = True, 
            boxprops=dict(facecolor='r'), medianprops = dict(color='k'))
plt.xlabel('Altitud (m)')
plt.title('Diagrama de caja de la Altitud de las ciudades de California');
```

*Juntamos los gráficos*

```python
# Datos secundarios necesarios para los gráficos:

# Ciudades con mayor población:
df_top_pop = df_casas.nlargest(5,'population_total').reset_index()
df_top_pop.population_total = df_top_pop.population_total / 1000
df_top_pop.city = df_top_pop.city.str.replace('LosAngeles','Los Ángeles')\
    .str.replace('SanDiego','San Diego')\
    .str.replace('SantaAna','Santa Ana')
colores = ['blue','grey','red','olive','brown']

# Extracción de outliers:
outliers = df_casas.nlargest(2,'area_water_km2')
coord_outliers = []
for ciudad in outliers.city:
    x = df_casas.loc[df_casas.city == ciudad,'area_water_km2']
    y = df_casas.loc[df_casas.city == ciudad,'population_total']/1000
    coord_outliers.append((float(x),float(y)))
coord_outliers
```

```python
fig, ax = plt.subplots(2,2, figsize = (20,20))

# Diagrama de barras de las ciudades más pobladas:

ax[0,0].bar(df_top_pop.city, height = df_top_pop.population_total, color = colores, alpha=0.5)
for i in range(len(df_top_pop)):
    ax[0,0].text(i, (df_top_pop.population_total[i]) +100, df_top_pop.population_total[i], ha = 'center')
ax[0,0].set_ylim(0,5000)
ax[0,0].set_ylabel('Habitantes (miles)')
ax[0,0].set_xlabel('Ciudades')
ax[0,0].set_title('Ciudades más pobladas de California');

# Histograma de la superficie de las ciudades en km2:

ax[0,1].hist(df_casas.area_total_km2, bins = 50, color = 'steelblue', label = 'Hist Area')
fuente = {'font':'Bradley Hand itc','size':18}
ax[0,1].set_xlabel('Superficie ($km^2$)',fontdict=fuente)
ax[0,1].set_ylabel('Número de ciudades',fontdict=fuente)
ax[0,1].legend();


# Diagrama de dispersión del área de agua de la ciudad en km2 vs habitantes por ciudad:

ax[1,0].scatter(df_casas.area_water_km2, df_casas.population_total / 1000, color = 'g', alpha=0.5)
ax[1,0].set_xlabel('Área de agua en la ciudad ($km^2$)')
ax[1,0].set_ylabel('Población (miles habitantes)')
ax[1,0].set_title('Población por kilómetro cuadrado de agua')
coordtext = (60,3500)
ax[1,0].annotate('Outliers',xy = coord_outliers[0], xytext=coordtext, arrowprops=dict(arrowstyle = '->'))
ax[1,0].annotate('Outliers',xy = coord_outliers[1], xytext=coordtext, arrowprops=dict(arrowstyle = '->'));

# Boxplot de la altitud de las ciudades:

ax[1,1].boxplot(df_casas.elevation_m, orientation = 'horizontal', patch_artist = True, 
            boxprops=dict(facecolor='r'), medianprops = dict(color='k'))
ax[1,1].set_xlabel('Altitud (m)')
ax[1,1].set_title('Diagrama de caja de la Altitud de las ciudades de California');
```
